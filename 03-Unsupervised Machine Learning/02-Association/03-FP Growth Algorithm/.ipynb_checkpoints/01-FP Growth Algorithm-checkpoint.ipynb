```python
#  importing libraries
from collections import defaultdict
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Node
class FPNode:
    def __init__(self, name, frequency, parent):
        self.name = name
        self.count = frequency
        self.parent = parent
        self.link = None
        self.children = {}

    def increment(self, frequency):
        self.count += frequency
```

```python
class FPTree:
    def __init__(self, transactions, threshold):
        self.frequent_itemsets = []
        self.header_table = defaultdict(list)
        self.threshold = threshold
        self.root = FPNode(None, 1, None)
        self.build_tree(transactions)

    def build_tree(self, transactions):
        item_counts = defaultdict(int)
        for transaction in transactions:
            for item in transaction:
                item_counts[item] += 1

        # Remove items below the threshold
        item_counts = {k: v for k, v in item_counts.items() if v >= self.threshold}

        # Build the FP-Tree
        for transaction in transactions:
            transaction = [item for item in transaction if item in item_counts]
            transaction.sort(key=lambda x: item_counts[x], reverse=True)
            if transaction:
                self.insert_tree(transaction, self.root)

    def insert_tree(self, items, node):
        first = items[0]
        if first in node.children:
            node.children[first].increment(1)
        else:
            node.children[first] = FPNode(first, 1, node)
            self.header_table[first].append(node.children[first])

        if len(items) > 1:
            self.insert_tree(items[1:], node.children[first])

    def mine_patterns(self):
        for item in sorted(self.header_table.keys(), key=lambda x: len(self.header_table[x])):
            self.frequent_itemsets.append(([item], sum(node.count for node in self.header_table[item])))
            conditional_pattern_base = []
            for node in self.header_table[item]:
                path = []
                parent = node.parent
                while parent.parent is not None:
                    path.append(parent.name)
                    parent = parent.parent
                conditional_pattern_base.extend([path] * node.count)

            conditional_tree = FPTree(conditional_pattern_base, self.threshold)
            conditional_tree.mine_patterns()
            for pattern, count in conditional_tree.frequent_itemsets:
                self.frequent_itemsets.append(([item] + pattern, count))
```

```python
#  visualization
def visualize_fp_tree(fp_tree):
    G = nx.DiGraph()
    labels = {}

    def add_edges(node, parent_name=None):
        if node.name:
            G.add_node(node.name, label=f'{node.name}:{node.count}')
            labels[node.name] = f'{node.name}:{node.count}'
        for child in node.children.values():
            if parent_name:
                G.add_edge(parent_name, child.name)
            add_edges(child, child.name)

    add_edges(fp_tree.root)
    pos = nx.spring_layout(G)
    nx.draw(G, pos, labels=labels, with_labels=True, node_size=3000, node_color='lightblue', font_size=10)
    plt.title("FP-Tree Visualization")
    plt.show()
```

```python
transactions = [['A', 'B', 'D', 'E'], 
                ['B', 'C', 'E'], 
                ['A', 'B', 'C', 'E'], 
                ['B', 'C'], 
                ['A', 'B', 'C', 'E']]

threshold = 2
fp_tree = FPTree(transactions, threshold)
fp_tree.mine_patterns()

print("Frequent Itemsets:")
for itemset, count in fp_tree.frequent_itemsets:
    print(f"Itemset: {itemset}, Count: {count}")

# Visualize the FP-Tree
visualize_fp_tree(fp_tree)
```
