```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 100)
true_slope = 2.5
true_intercept = 1.0
true_sigma = 1.0
y = true_slope * X + true_intercept + np.random.normal(0, true_sigma, size=X.shape)
```

```python
plt.scatter(X, y, label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import BayesianRidge

# Prepare the data
X = X.reshape(-1, 1)

model = BayesianRidge()
model.fit(X, y)

# Print the estimated parameters
print(f'Estimated intercept: {model.intercept_}')
print(f'Estimated slope: {model.coef_[0]}')
```

```python
# Generate predictions
y_pred = model.predict(X)

# Plot the fitted regression line
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='red', label='Bayesian Ridge Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
