```python
#  importing libraries
import matplotlib.pyplot as plt
```

```python
# Example frequent itemsets and their support counts
itemsets = ['A', 'B', 'C', 'AB', 'AC', 'BC', 'ABC']
support_counts = [4, 4, 4, 3, 3, 3, 2]
```

```python
plt.bar(itemsets, support_counts, color='blue')
plt.xlabel('Itemsets')
plt.ylabel('Support Count')
plt.title('Frequent Itemsets with Support Counts')
plt.show()
```
