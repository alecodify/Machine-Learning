```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 5)
true_coeffs = np.array([1.5, -2, 0, 3, 0])
y = X @ true_coeffs + np.random.normal(0, 0.5, 100)
```

```python
# Plot the original data
plt.scatter(X[:, 0], y, label='Data')
plt.xlabel('Feature 1')
plt.ylabel('y')
plt.legend()
plt.show()
```

### Lasso Regression Model

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)  # alpha is the regularization parameter (λ)
lasso.fit(X, y)

# Print the estimated coefficients
print(f'Estimated coefficients (Lasso): {lasso.coef_}')
```

### Ridge Regression Model

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1)  # alpha is the regularization parameter (λ)
ridge.fit(X, y)

# Print the estimated coefficients
print(f'Estimated coefficients (Ridge): {ridge.coef_}')
```

```python
# Plot the true coefficients vs. estimated coefficients
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(true_coeffs, 'o-', label='True Coefficients')
plt.plot(lasso.coef_, 'x-', label='Lasso Coefficients')
plt.xlabel('Feature Index')
plt.ylabel('Coefficient Value')
plt.title('Lasso Regression')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(true_coeffs, 'o-', label='True Coefficients')
plt.plot(ridge.coef_, 'x-', label='Ridge Coefficients')
plt.xlabel('Feature Index')
plt.ylabel('Coefficient Value')
plt.title('Ridge Regression')
plt.legend()

plt.tight_layout()
plt.show()
```
