```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
dataset = pd.read_csv("01-Density Based Clustering.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import StandardScaler
# Standardize the features
ss = StandardScaler()
data_scaled = ss.fit_transform(dataset)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(data_scaled)
```

```python
# Add the labels to the original dataframe
dataset['Cluster'] = labels
```

```python
from sklearn.decomposition import PCA
# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)
```

```python
# Create a DataFrame with the PCA results and cluster labels
df_pca = pd.DataFrame(data_pca, columns=['PCA1', 'PCA2'])
df_pca['Cluster'] = labels
```

```python
import seaborn as sns
# Plot the PCA results
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='Cluster', palette='Set1', s=100)
plt.title('DBSCAN Clustering (PCA Reduced Dimensions)')
plt.show()
```
