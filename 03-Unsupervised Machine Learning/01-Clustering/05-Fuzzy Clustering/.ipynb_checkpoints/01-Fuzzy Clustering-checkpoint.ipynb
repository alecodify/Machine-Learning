```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# getting data
dataset = pd.read_csv("01-Fuzzy Clustering.csv")
dataset.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(dataset)
```

```python
from fcmeans import FCM
# Specify the number of clusters, for example 3
fcm = FCM(n_clusters=3)
fcm.fit(data_scaled)
```

```python
# Retrieve the cluster centers and the membership grades
fcm_centers = fcm.centers
fcm_labels = fcm.u.argmax(axis=1)
```

```python
# Add the cluster labels to the dataframe
dataset['cluster'] = fcm_labels
```

```python
# Plot the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=dataset, x=dataset.columns[0], y=dataset.columns[1], hue='cluster', palette='viridis', s=100, alpha=0.7)
plt.scatter(fcm_centers[:, 0], fcm_centers[:, 1], c='red', s=300, marker='x')  # plot the centers
plt.title('Fuzzy C-Means Clustering')
plt.xlabel(dataset.columns[0])
plt.ylabel(dataset.columns[1])
plt.show()
```
